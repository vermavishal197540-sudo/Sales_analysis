# **Exploratory** **Data** **Analysis**

 **Segment**

```python
# plotting a bar chart for segment and it's count
ax = sns.countplot(x = 'Segment',hue='Segment',data = df, palette=['red', 'green','blue'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for segment and it's count
ax = sns.countplot(x = 'Segment',data = df, hue='Segment', palette=['red', 'green','blue'])
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the Sales are consumer and even the purchasing power of consumer are greater than *home office* *italicized text*

## **Region**

```python
ax = sns.countplot(data = df, x = 'Country', hue = 'Region', palette=['red', 'green','blue','yellow'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_region = df.groupby(['Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
sns.barplot(x = 'Region',y= 'Sales' ,data = sales_region, palette='tab20',hue='Region')
```

From above graphs we can see that most of the sales are from west region

```python
# total number of Sales from region
sales_region = df.groupby(['Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_region, x = 'Region',y= 'Sales', palette='plasma',hue='Region')
```

```python
# total amount/sales from top 10 city
sales_city = df.groupby(['City'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_city, x = 'City',y= 'Sales', palette=['red', 'yellow', 'pink', 'green', 'blue', 'purple', 'orange', 'brown', 'gray', 'cyan'],hue='City')
```

From above graphs we can see that most of the sales & total sales/amount are from NY City,Los Angeles,Seattle.

# **CATEGORY**

```python
ax = sns.countplot(data = df, x = 'Category', palette=['red', 'green','blue'],hue='Category')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_sub_category = df.groupby(['Sub_Category', 'Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_sub_category, x = 'Sub_Category',y= 'Sales', hue='Region', palette=[ 'red', 'green','blue','yellow'])
```

From above graphs we can see that most of the sales are chairs 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Category', palette='viridis',hue='Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_category = df.groupby(['Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_category, x = 'Category',y= 'Sales', palette='plasma',hue='Category')
```

From above graphs we can see that most of the sales are from  IT sector

# **Sub** **Category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Sub_Category', palette='tab10',hue='Sub_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sub_category = df.groupby(['Sub_Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sub_category, x = 'Sub_Category',y= 'Sales', palette=['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan'],hue='Sub_Category')
```

From above graphs we can see that most of the sold products are from Phones, Chairss and Electronics category

```python
orders = df.groupby(['Order_ID'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = orders, x = 'Order_ID',y= 'Sales', palette='hsv',hue='Order_ID')
```
